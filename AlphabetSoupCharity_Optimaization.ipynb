## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
print("\nThe target variables are the values found specificaly within the 'IS_SUCCESSFUL' column.")
print("Thus, the feature variables are the values found in all other columns of importance (i.e. columns other than the 'IS_SUCCESSFUL' column).\n")
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

# List of columns to drop
columns_to_drop = ['EIN', 'NAME']

# Drop columns only if they exist in the DataFrame  (NOTE: this code can be re-run without erroring out)
application_df.drop(columns=[col for col in columns_to_drop if col in application_df.columns], inplace=True)

application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`



cutoff = 500

# Get the value counts of APPLICATION_TYPE
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()

# Filter the values that have counts below the cutoff
application_types_to_replace = application_type_counts[application_type_counts < cutoff].index.tolist()
print(f"application_types_to_replace: (replace these as 'Other')\n{application_types_to_replace}\n")



# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
#################################
# SPECIAL PANDAS DISPLAY OPTION #
#################################
# Set the maximum number of rows to display (None = display all rows)
pd.set_option('display.max_rows', None)
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"

application_df['CLASSIFICATION'].value_counts()
```

```python
#################################
# SPECIAL PANDAS DISPLAY OPTION #
#################################
# Reset to the default max rows value (usually 10)
pd.reset_option('display.max_rows')
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1

# Get the value counts for the 'CLASSIFICATION' column
classification_counts = application_df['CLASSIFICATION'].value_counts()

# Filter the value counts to only display those greater than 1
filtered_counts = classification_counts[classification_counts > 1]

# Display the filtered counts
print(filtered_counts)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`



cutoff = 1000

# Get the value counts of APPLICATION_TYPE
classification_type_counts = application_df['CLASSIFICATION'].value_counts()

# Filter the values that have counts below the cutoff
classifications_to_replace = classification_type_counts[classification_type_counts < cutoff].index.tolist()
print(f"classifications_to_replace: (replace these as 'Other')\n{classifications_to_replace}\n")

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
#################################
# SPECIAL PANDAS DISPLAY OPTION #
#################################
# Set the maximum number of columns to display (None = display all columns)
pd.set_option('display.max_columns', None)
```

```python
print("########## Before Converting Categorical Data to Numeric ##########")
application_df.head()
 
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)
```

```python
print("\n\n########## After Converting Categorical Data to Numeric ##########")
application_df.head()
```

```python
#################################
# SPECIAL PANDAS DISPLAY OPTION #
#################################
# Reset the display.max_columns option to its default value
pd.reset_option('display.max_columns')
```

```python
# Split our preprocessed data into our features and target arrays
y_labels_df = application_df['IS_SUCCESSFUL']                     # y - labels - label vector - target variables (dependent variables)
x_features_df = application_df.drop(columns=['IS_SUCCESSFUL'])    # X - features - feature matrix - input variables (independent variables)

# Split the preprocessed data into a training and testing dataset
test_data_percentage = (20/100)    # 20% of the data will be used for testing, and 80% for training. (FYI: train_test_split default is 25%)
X_train, X_test, y_train, y_test = train_test_split(x_features_df, y_labels_df, test_size=test_data_percentage, random_state=1)
```

```python
print("Before split")
print(f"x_features_df:  {x_features_df.shape}")
print(f"y_labels_df:    {y_labels_df.shape}")

print()
print("After split")
print(f"X_train:        {X_train.shape}")
print(f"y_train:        {y_train.shape}")
print(f"X_test:         {X_test.shape}")
print(f"y_test:         {y_test.shape}")

```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
print("After scaling")
print(f"X_train_scaled:  {X_train_scaled.shape}")
print(f"X_test_scaled:   {X_test_scaled.shape}")
```

```python
print("########## After Scaling the Data ##########\n")

# Number of rows you want to view (similar to head())
num_rows = 2

# Print the first few rows
print("Scaled Training Data\n---------------------")
print(X_train_scaled[:num_rows])

# Print the first few rows
print("\nScaled Testing Data\n--------------------")
print(X_test_scaled[:num_rows])
```

```python
# Get the number of columns ( to use as input to the ML Model's Input Layer )
columns_in_training_data = X_train_scaled.shape[1]    # 43
print(f"Input Layer: {columns_in_training_data} columns\n")
```

# OPTIMIZATION 1
## Add More Neurons to a Hidden Layer

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

# Create the Keras Sequential model
nn = tf.keras.models.Sequential()

# Add Input layer
nn.add(tf.keras.layers.Input(shape=(columns_in_training_data,)))

# Add First Dense layer ( with more neurons ... increased from 80 to 160 )
nn.add(tf.keras.layers.Dense(units=160, activation="relu"))

# Add Second Dense layer ( with more neurons ... increased from 30 to 60 )
nn.add(tf.keras.layers.Dense(units=60, activation="relu"))

# Add Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the Sequential model
nn.summary()


# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])


# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

# Export model to HDF5 file
nn.save("AlphabetSoupCharity_Optimaization_1.h5", save_format='h5')
```

# OPTIMIZATION 2
## Add More Hidden Layers

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.


# Create the Keras Sequential model
nn = tf.keras.models.Sequential()

# Add Input layer
nn.add(tf.keras.layers.Input(shape=(columns_in_training_data,)))


# ( ... With More Hidden layers )

# Add First Dense layer
nn.add(tf.keras.layers.Dense(units=80, activation="relu"))

# Add Second Dense layer
nn.add(tf.keras.layers.Dense(units=60, activation="relu"))

# Add Third Layer
nn.add(tf.keras.layers.Dense(units=40, activation="relu"))

# Add Fourth Layer
nn.add(tf.keras.layers.Dense(units=20, activation="relu"))


# Add Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the Sequential model
nn.summary()


# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])


# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimaization_2.h5", save_format='h5')
```
